# 4. Outlier Detection

### Outlier Detection by Site

Let's use statistical methods to identify sites with abnormal behavior in terms of energy consumption.


```python
#Outlier detection by site using the IQR (Interquartile Range) method

print("Outlier detection by site")

# calculate Q1, Q3 and IQR for average consumption by site
Q1 =site_stats['Average consumption'].quantile(0.25)
Q3 =site_stats['Average consumption'].quantile(0.75)
IQR =Q3- Q1

#define thresholds for outliers
lower_bound= Q1 -1.5 *IQR
upper_bound=Q3 + 1.5 * IQR

print(f"\nStatistics for outlier detection:")
print(f"Q1 (25th percentile): {Q1:.2f} kWh")
print(f"Q3 (75th percentile):{Q3:.2f} kWh")
print(f"IQR: {IQR:.2f} kWh")
print(f"Lower bound: {lower_bound:.2f} kWh")
print(f"Upper bound:{upper_bound:.2f} kWh")

# identify outlier sites
site_stats['Is_Outlier'] =(site_stats['Average consumption'] < lower_bound) | (site_stats['Average consumption'] > upper_bound)
outlier_sites = site_stats[site_stats['Is_Outlier']]

print(f"\noutlier sites detected :{len(outlier_sites)} site(s)")
if len(outlier_sites) >0 :

    print("\nDetails of outlier sites:")
    print(outlier_sites[['Average consumption', 'Median consumption', 'Standard deviation', 'Number of records']])

    # classify outliers
    high_outliers =outlier_sites[outlier_sites['Average consumption']> upper_bound]
    low_outliers = outlier_sites[outlier_sites['Average consumption'] < lower_bound]

    if len(high_outliers)> 0: 
        print(f"\nSites with high abnormal consumption:{list(high_outliers.index)}")
    if len(low_outliers) >0:
        print(f"\nSites with low abnormal consumption: {list(low_outliers.index)}")
else:
    print("No outlier sites detected with the IQR method")

# visualization
fig, ax= plt.subplots(figsize=(14, 6))
colors = ['red' if outlier else 'steelblue' for outlier in site_stats['Is_Outlier']]
bars = ax.bar(site_stats.index.astype(str), site_stats['Average consumption'], color=colors, edgecolor='black')

ax.axhline(y=upper_bound, color='red', linestyle='--', linewidth=2, label=f'Upper bound ({upper_bound:.0f} kWh)')
ax.axhline(y=lower_bound, color='orange', linestyle='--', linewidth=2, label=f'Lower bound ({lower_bound:.0f} kWh)')
ax.axhline(y=site_stats['Average consumption'].median(), color='green', linestyle='-', linewidth=2, label=f'Median ({site_stats["Average consumption"].median():.0f} kWh)')

ax.set_title('Outlier site detection :average consumption', fontsize=16, fontweight='bold')
ax.set_xlabel('Site ID', fontsize=12)
ax.set_ylabel('Average consumption (kWh)', fontsize=12)
ax.legend(fontsize=11)
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

# 6. Data preprocessing and feature engineering


Completing the missing values

```python
weather_new = weather_df.copy()

# make sure timestamp is datetime
weather_new["timestamp"] = pd.to_datetime(weather_new["timestamp"])

#add month and day
weather_new["month"] = weather_new["timestamp"].dt.month
weather_new["day"]   = weather_new["timestamp"].dt.day

# columns where we want to fill missing values
weather_cols = [
    "air_temperature",
    "cloud_coverage",
    "dew_temperature",
    "precip_depth_1_hr",
    "sea_level_pressure",
    "wind_direction",
    "wind_speed"]

#fill the NaN with monthly mean for each site
weather_new[weather_cols] = (
    weather_new
    .groupby(["site_id", "month"])[weather_cols]
    .transform(lambda x: x.fillna(x.mean())))

#for the remaining NaN, we use the mean of each site
for i in weather_cols:
    weather_new[i] = (
        weather_new
        .groupby("site_id")[i]
        .transform(lambda x: x.fillna(x.mean())))

    # if there are still NaN, we use global mean of the column
    global_mean = weather_new[i].mean()
    weather_new[i].fillna(global_mean, inplace=True)

# remove the columns month, day
weather_new.drop(columns=["month", "day"], inplace=True)

# check that nothing is missing
print(weather_new.isnull().sum())
```

Time feature engineering

```python
#function that adds hour, day, weekday and month so the model can understand how energy use changes over time
def add_time_features(df):
    # we make sure timestamp is datetime
    df["timestamp"] = pd.to_datetime(df["timestamp"])

    # create new time features
    df["hour"] = df["timestamp"].dt.hour
    df["day"] = df["timestamp"].dt.day
    df["weekday"] = df["timestamp"].dt.weekday
    df["month"] = df["timestamp"].dt.month
    return df
```

Building Age

we saw that the variable year_built alone was not very useful. However, we can use it to calculate the age of the building, which may impact energy consumption.

```python
# we create a new column for building age
#subtract the construction year from 2016 because the dataset only contains data from 2016, so we calculate the building age at that moment
building_metadata_df["building_age"] = 2016 - building_metadata_df["year_built"]

# replace negative or missing ages with the median age
median_age = building_metadata_df["building_age"].median()
building_metadata_df["building_age"].fillna(median_age, inplace=True)
```

Encoding categorical columns

```python
from sklearn.preprocessing import LabelEncoder

# we create a label encoder
encoder = LabelEncoder()
#convert primary_use text into numbers
building_metadata_df["primary_use_encoded"] = encoder.fit_transform(
    building_metadata_df["primary_use"])
print(building_metadata_df[["primary_use", "primary_use_encoded"]].head(20))
```

the text categories in "primary_use" were changed into numbers, so the model can work with them. For example, "education" became 0, "entertainment/public assembly" became 1, "lodging/residential" became 4, "office" became 6

Feature selection : choosing input & target variables

```python
#we select the columns that will be useful for prediction
# we keep weather data, building info and time features
selected_features = [
    "building_id",
    "site_id",
    "meter",
    "building_age",
    "square_feet",
    "primary_use_encoded",
    "air_temperature",
    "dew_temperature",
    "wind_speed",
    "sea_level_pressure",
    "hour",
    "day",
    "weekday",
    "month"]

# target variable (what we want to predict)
target = "meter_reading"

print("Features selected for the model:")
print(selected_features)
```

We selected these features because they were the most important in our EDA. They describe the building, the weather, and the time  which are the three main factors that influence energy consumption.

```python
# root mean squared error
def rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred) ** 2))
```

Tune hyperparameters

we tune hyperparameters to make the model perform better. The default settings aren't optimal, so testing different values helps to improve the predictions

```python
import xgboost as xgb

#set the default parameters
n_estimators_base    = 300 #number of trees
learning_rate_base   = 0.05
max_depth_base       = 6
subsample_base       = 0.8 #use 80% of the rows for each tree
colsample_bytree_base = 0.8  # use 80% of the features for each tree
objective_base       = "reg:squarederror"
tree_method_base     = "hist"
random_state_base    = 42
n_jobs_base          = -1

print("default parameters :")
print("n_estimators     :", n_estimators_base)
print("learning_rate    :", learning_rate_base)
print("max_depth        :", max_depth_base)
print("subsample        :", subsample_base)
print("colsample_bytree :", colsample_bytree_base)
```

Tuning of n_estimators

```python
# create a list of values to test
n_estimators_list = [200, 300, 400]

best_n_estimators = None  #to store the best value
best_val_rmse_n   = float("inf") #start with a big rmse

for n_estimators in n_estimators_list:
    print("\ntest with n_estimators =", n_estimators)

    # create the model
    model = xgb.XGBRegressor(
        n_estimators=n_estimators,
        learning_rate=learning_rate_base,
        max_depth=max_depth_base,
        subsample=subsample_base,
        colsample_bytree=colsample_bytree_base,
        objective=objective_base,
        tree_method=tree_method_base,
        random_state=random_state_base,
        n_jobs=n_jobs_base)

    # train the model
    model.fit(X_train_xgb, y_train_xgb)

    #predictions
    y_train_pred = model.predict(X_train_xgb)
    y_val_pred   = model.predict(X_val_xgb)

    # rmse
    rmse_train = rmse(y_train_xgb, y_train_pred)
    rmse_val   = rmse(y_val_xgb, y_val_pred)

    print("Train rmse :", rmse_train)
    print("Val   rmse :", rmse_val)

    # keep the best value
    if rmse_val < best_val_rmse_n:
        best_val_rmse_n   = rmse_val
        best_n_estimators = n_estimators

print("\nbest n_estimators", best_n_estimators,
      "with validation rmse =", best_val_rmse_n)

# update the base value
n_estimators_base = best_n_estimators
```

Interpretation :

When n_estimators = 200, the validation rmse is 1.416, which is the worst result

With n_estimators = 300, the validation rmse improves to 1.370.

When n_estimators = 400, the validation rmse becomes 1.330, which is the best score.
So, it gives the lowest validation error, meaning the model predicts better with this value, n_estimators = 400 is the best choice.

Tuning of max_depth

```python
# values to test
max_depth_list = [4, 6, 8]

best_max_depth = None
best_val_rmse_d = float("inf")

for max_depth in max_depth_list:
    print("\nTest with max_depth =", max_depth)

    model = xgb.XGBRegressor(
        n_estimators=n_estimators_base,
        learning_rate=learning_rate_base,
        max_depth=max_depth,
        subsample=subsample_base,
        colsample_bytree=colsample_bytree_base,
        objective=objective_base,
        tree_method=tree_method_base,
        random_state=random_state_base,
        n_jobs=n_jobs_base)

    model.fit(X_train_xgb, y_train_xgb)

    y_train_pred = model.predict(X_train_xgb)
    y_val_pred   = model.predict(X_val_xgb)

    rmse_train = rmse(y_train_xgb, y_train_pred)
    rmse_val   = rmse(y_val_xgb, y_val_pred)

    print("Train rmse :", rmse_train)
    print("val   RMSrmseE :", rmse_val)

    if rmse_val < best_val_rmse_d:
        best_val_rmse_d = rmse_val
        best_max_depth  = max_depth

print("\nbest max_depth =", best_max_depth,
      "with validation rmse =", best_val_rmse_d)

max_depth_base = best_max_depth
```

Interpretation :

with max_depth = 4 :the rmse is high with train and validation.

When max_depth = 6 : the results are better

with max_depth = 8 :the model gives the lowest validation rmse  (1.23)

So, max_depth = 8 is the best choice because it gives the smallest validation error.



Optimized Final XGBoost Model

```python
#XGBoost model with tuned hyperparameters
xgb_best_model = xgb.XGBRegressor(
    n_estimators=n_estimators_base,
    learning_rate=learning_rate_base,
    max_depth=max_depth_base,
    subsample=subsample_base,
    colsample_bytree=colsample_bytree_base,
    objective=objective_base,
    tree_method=tree_method_base,
    random_state=random_state_base,
    n_jobs=n_jobs_base)

xgb_best_model.fit(X_train_xgb, y_train_xgb)

# predictions
y_train_pred_best = xgb_best_model.predict(X_train_xgb)
y_val_pred_best   = xgb_best_model.predict(X_val_xgb)

rmse_train_best = rmse(y_train_xgb, y_train_pred_best)
rmse_val_best   = rmse(y_val_xgb, y_val_pred_best)

print("\nXGBoost optimized - train rmse :", rmse_train_best)
print("XGBoost optimized - validation rmse :", rmse_val_best)

# R²
r2_train_best = r2_score(y_train_xgb, y_train_pred_best)
r2_val_best   = r2_score(y_val_xgb,   y_val_pred_best)

print("XGBoost optimized  - train R² :", r2_train_best)
print("XGBoost optimized - val   R² :", r2_val_best)
```

Interpretation:

- train RMSE = 0.92 : the model makes small errors on the training data

- validation RMSE = 1.23 :the model also performs well on new data
- train R² = 0.82 ,the model explains 82% of the training data
- and validation R² = 0.65:the model explains 65% of the validation data

So, the optimized XGBoost model gives better results than the previous models.
